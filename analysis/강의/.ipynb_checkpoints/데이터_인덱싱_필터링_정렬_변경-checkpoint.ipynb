#### 2. 열(Column) 추가/제거

```python
# 열 선택
df_cyl = df['cyl'] # == df.cyl
df_cyl.head(3)
```

```python
# 다중 열을 가져오려면
df_new = df[['cyl', 'mpg']]
df_new.head(3)
```

```python
# 열 제거
# df.head(3)
df.drop(columns=['car', 'mpg', 'cyl']).head(3)
```

```python
# 열 추가
df2 = df.copy() # 이렇게 복사해야 함 ( df2 = df 복사는 df가 바뀌면 df2도 바뀜 )
df2['new'] = df['mpg'] + 10
df2.head(3)
```

#### 3) 데이터 필터링

```python
# 1개 조건 필터링
# cyl=4인 데이터의 수
cond1 = ( df['cyl']==4 )
len(df[cond1])

# cyl_4 = df[df['cyl'] == 4 ]
# print(len(cyl_4))
```

```python
# mpg가 22이상인 데이터 수
cond2 = (df['mpg'] >= 22)
print(len(df[cond2]))
```

```python
df[df['mpg']>= 22]
```

```python
# 2개 조건 필터링
df[cond1 & cond2]
```

```python
print(len(df[cond1 & cond2]))
```

```python
# 2개 조건 필터링 후 데이터 개수 (or)
df[cond1 | cond2]
```

```python
print(len( df[cond1 | cond2] ))
```

```python
# 한번에 코딩 할 경우
print(len(df[ (df['cyl']==4) & (df['mpg'] >= 22) ] ))
print(len(df[ (df['cyl']==4) | (df['mpg'] >= 22) ] ))
```

#### 4) 데이터 정렬

```python
# 내림차순 정렬
df.sort_values('mpg', ascending=False).head()
```

```python
# 오름차순 정렬
df.sort_values('mpg', ascending=True).head()
```

#### 5) 데이터 변경 ( 조건문 )

```python
import numpy as np
df = pd.read_csv("../mtcars.csv")

# np.where 활용
# hp 변수 값 중에서 205가 넘는 값은 205로 처리하고 나머지는 그대로 유지
df['hp'] = np.where(df['hp']>= 205, 205, df['hp'])

# 내림차순 정렬
df.sort_values('hp', ascending=False).head(10)

# 활용 : 이상치를 Max 값이나 Min 값으로 대체 할 경우 조건 문 활용
```
