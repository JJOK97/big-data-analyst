```python
###############  실기환경 복사 영역  ###############
import pandas as pd
import numpy as np
# 실기 시험 데이터셋으로 셋팅하기 (수정금지)
from sklearn.datasets import load_iris
# Iris 데이터셋을 로드
iris = load_iris()
x = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
y = iris.target   # 'setosa'=0, 'versicolor'=1, 'virginica'=2
y = np.where(y>0, 1, 0) # setosa 종은 0, 나머지 종은 1로 변경

# 실기 시험 데이터셋으로 셋팅하기 (수정금지)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, 
                                                    stratify=y,
                                                    random_state=2023)

x_test = pd.DataFrame(x_test)
x_train = pd.DataFrame(x_train)
y_train = pd.DataFrame(y_train)
y_train.columns = ['species']

# 결측치 삽입
x_test.loc[x_test.index[0], 'sepal_length'] = None
x_train.loc[x_train.index[0], 'sepal_length'] = None

# 이상치 삽입
x_train.loc[x_train.index[0], 'sepal_width'] = 150
###############  실기환경 복사 영역  ###############
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train['sepal_length'] = x_train['sepal_length'].fillna(x_train['sepal_length'].mean())
x_test['sepal_length'] = x_test['sepal_length'].fillna(x_test['sepal_length'].mean())
```

```python
Q1 = x_train.quantile(0.25)
Q3 = x_train.quantile(0.75)
IQR = Q3 - Q1
outliers = ((x_train < Q1 - 1.5 * IQR) | (x_train > Q3 + 1.5 * IQR))
outliers.sum()
```

```python
Q1 = x_train['sepal_width'].quantile(0.25)
Q3 = x_train['sepal_width'].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

x_train['sepal_width'] = np.where(x_train['sepal_width'] < lower, lower, x_train['sepal_width'])
```

```python
x_train['sepal_width'] = np.where(x_train['sepal_width'] > upper, upper, x_train['sepal_width'])
```

```python
x_test['sepal_width'] = np.where(x_test['sepal_width'] < lower, lower, x_test['sepal_width'])
x_test['sepal_width'] = np.where(x_test['sepal_width'] > upper, upper, x_test['sepal_width'])
```

```python
Q1 = x_train['sepal_width'].quantile(0.25)
Q3 = x_train['sepal_width'].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

outliers = ((x_train['sepal_width'] < lower) | (x_train['sepal_width'] > upper))
outliers.sum()
```

```python
print(y_train.value_counts())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train['species'], test_size = 0.2, stratify = y_train['species'], random_state = 2025)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state = 2025)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_val)
```

```python
from sklearn.metrics import f1_score, roc_auc_score

f1 = f1_score(y_val, y_pred)
auc = roc_auc_score(y_val, y_pred)
```
