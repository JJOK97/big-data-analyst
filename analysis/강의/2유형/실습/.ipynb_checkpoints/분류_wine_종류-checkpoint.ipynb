```python
Q1 = x_train.quantile(0.25)
Q3 = x_train.quantile(0.75)
IQR = Q3 - Q1
outliers = ((x_train < (Q1 - 1.5 * IQR)) | (x_train > (Q3 + 1.5 * IQR))).sum()
print(outliers)
```

```python
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

def clip(s):
    return np.where(s < lower[s.name], lower[s.name], np.where(s > upper[s.name], upper[s.name], s))
```

```python
x_train = x_train.apply(clip, axis = 0)
x_test = x_test.apply(clip, axis = 0)
```

Ïó¨Í∏∞ÍπåÏßÄ Ï†ÑÏ≤òÎ¶¨ Î∞è Î∂ÑÎ¶¨Í∞Ä ÎÅùÏù¥ÎÇ¨Í≥†, Î™®Îç∏ÎßÅÏùÑ Ìï¥ÏïºÌï®

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train['target'], test_size = 0.2, stratify = y_train['target'], random_state = 2025) 
```

```python
print(x_train.head(3))
print(x_val.head(3))
print(y_train.head(3))
print(y_val.head(3))
```

```python
print(x_train.dtypes)
print(x_val.dtypes)
print(y_train.dtypes)
print(y_val.dtypes)
```

```python
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
print(x_train.info)
print(x_val.info)
print(y_train.info)
print(y_val.info)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 2025)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_val)
```

Ïù¥Î†áÍ≤åÌïòÎ©¥ ÌïôÏäµÏù¥ ÎÅùÎÇ¨Í≥† Í∑∏Îã§ÏùåÏóê Í≤ÄÏ¶ùÏùÑ Ìï¥ÏïºÌï®

Î∂ÑÎ•òÎ™®Îç∏Ïùò ÏÑ±Îä•ÌèâÍ∞ÄÌï¥ÏïºÌï®. Î∂ÑÎ•òÏóêÏÑúÎäî Ï†ïÌôïÎèÑ, F1 Score, AUCÏùò Îì±ÏúºÎ°ú ÌèâÍ∞ÄÌïúÎã§.

```python
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score
```

```python
acc = accuracy_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred, average = 'macro')
# auc = roc_auc_score(y_val, y_pred)
```

```python
print(acc, '\n')
print(f1, '\n')
```
