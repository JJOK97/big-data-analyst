```python
Q1 = x_train.quantile(0.25)
Q3 = x_train.quantile(0.75)
IQR = Q3 - Q1
outliers = ((x_train < (Q1 - 1.5 * IQR)) | (x_train > (Q3 + 1.5 * IQR))).sum()
print(outliers)
```

```python
# Q1 = x_test.quantile(0.25)
# Q3 = x_test.quantile(0.75)
# IQR = Q3 - Q1
outliers = ((x_test < (Q1 - 1.5 * IQR)) | (x_test > (Q3 + 1.5 * IQR))).sum()
print(outliers)

# test 데이터도 train의 IQR로 계산해야함
```

```python
x_train['malic_acid'] = np.where(x_train['malic_acid'] < (Q1 - 1.5 * IQR)['malic_acid'], (Q1 - 1.5 * IQR)['malic_acid'], x_train['malic_acid'])
x_train['malic_acid'] = np.where(x_train['malic_acid'] > (Q3 + 1.5 * IQR)['malic_acid'], (Q3 + 1.5 * IQR)['malic_acid'], x_train['malic_acid'])
```

```python
Q1 = x_train.quantile(0.25)
Q3 = x_train.quantile(0.75)
IQR = Q3 - Q1
outliers = ((x_train < (Q1 - 1.5 * IQR)) | (x_train > (Q3 + 1.5 * IQR))).sum()
print(outliers)
```

```python
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

def clip(s):
    return np.where(s < lower[s.name], lower[s.name], np.where(s > upper[s.name], upper[s.name], s))
```

```python
x_train = x_train.apply(clip, axis = 0)
x_test = x_test.apply(clip, axis = 0)
```

여기까지 결측치 확인, 이상치 제거했고 변수처리 해줘야함. **파생변수 생성이나 원-핫 인코딩**

여기까지 전처리 및 분리가 끝이났고, 모델링을 해야함
좋은 모델링을 위해 validation 활용

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train['target'], test_size = 0.2, stratify = y_train['target'], random_state = 2025) 
```

```python
print(x_train.head(3))
print(x_val.head(3))
print(y_train.head(3))
print(y_val.head(3))
```

```python
print(x_train.info)
print(x_val.info)
print(y_train.info)
print(y_val.info)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 2025)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_val)
```

이렇게하면 학습이 끝났고 그다음에 검증을 해야함

분류모델의 성능평가해야함. 분류에서는 정확도, F1 Score, AUC의 등으로 평가한다.

```python
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score
```

```python
acc = accuracy_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred, average = 'macro')
# auc = roc_auc_score(y_val, y_pred)
```

```python
print(acc, '\n')
print(f1, '\n')
```
